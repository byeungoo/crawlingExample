```python
import requests
from bs4 import BeautifulSoup  #bs4라이브러리에서 BeautifulSoup 클래스를 사용하겠다.
```

```python
res = requests.get('http://v.media.daum.net/v/20170615203441266') #응답으로 html을 받아옴 (res.content)
#res.content
```

```python
soup = BeautifulSoup(res.content, 'html.parser')  # 문자열을 분석하는 것을 파싱이라고 합니다. BeautifulSoup를 사용하면 html5를 바로 파싱할 수 있습니다.
                                                  # 분석할 html파일을 첫번째 인자, 두번째 인자로 여러가지 파서가 있는데 일반적으로 사용하는 html.parser를 사용하도록 지정합니다.
                                                  # soup에는 분석한 데이터가 들어가 있습니다. (html의 트리구조라던가)
```

```python
mydata = soup.find('title') #함수로 원하는 부분을 지정하면 됨
```
